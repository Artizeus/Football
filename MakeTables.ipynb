```python
import pandas as pd
import browser_cookie3
import requests
import numpy as np
```

```python
COOKIE_DATA = browser_cookie3.chrome()
```

```python
def create_full_df(url, cookies):
    PAGE_OFFSET_INCR = 200
    offset = 0
    df = pd.DataFrame()
    for x in range(5):
        try:
            url_offset = url + '&offset='+str(offset)
            response = requests.get(url_offset, cookies=COOKIE_DATA)
            temp_df = pd.read_html(response.text)[0]
            df = pd.concat([df, temp_df], ignore_index=True, axis=0)
            offset += PAGE_OFFSET_INCR
        except:
            pass
    return df
```

```python
def initial_table_clean (dataframe):
    # Might have to make exception for when not 2 levels
    try:
        dataframe.columns = dataframe.columns.droplevel()
    except:
        pass
    dataframe.drop(dataframe.columns[[0]], axis=1, inplace=True)
    dataframe = dataframe.loc[:,~dataframe.columns.duplicated()].copy()
    if ('Unnamed: 7_level_1' in dataframe):
        dataframe = dataframe.rename(columns={'Unnamed: 7_level_1' : 'Home'})
    else:
        dataframe = dataframe.rename(columns={'Unnamed: 7' : 'Home'})
    dataframe['Home'] = np.where(dataframe['Home'] == '@', 1, 0)
    
    colname = (list(dataframe.columns))[2]
    colmove = dataframe.pop(colname)
    dataframe.insert(8, colmove.name, colmove)
    return dataframe
```

```python
offList = []
```

```python
url = 'https://stathead.com/football/team-game-finder.cgi?request=1&order_by=points&year_min=2022&year_max=2022'
GAME_POINTS = initial_table_clean(create_full_df(url, COOKIE_DATA))
GAME_POINTS.drop(['Pts.1'], axis=1, inplace=True)
display(GAME_POINTS)
offList.append(GAME_POINTS)
GAME_POINTS.name = 'GAME_POINTS'
```
