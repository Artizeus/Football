```python
url = 'https://stathead.com/football/team-game-finder.cgi?request=1&order_by=points&year_min=2022&year_max=2022&offset=0'
response = requests.get(url, cookies=COOKIE_DATA)
df = pd.read_html(response.text)[0]
display(df)
```

```python
PAGE_OFFSET_INCR = 200
def create_full_df(url=url, cookies=COOKIE_DATA):
    offset = 0
    df = pd.DataFrame()
    for x in range(5):
        try:
            url_offset = url + '&offset='+str(offset)
            response = requests.get(url_offset, cookies=COOKIE_DATA)
            temp_df = pd.read_html(response.text)[0]
            df = pd.concat([df, temp_df], ignore_index=True, axis=0)
            offset += PAGE_OFFSET_INCR
        except:
            pass
    return df
```

```python
url = 'https://stathead.com/football/team-game-finder.cgi?request=1&order_by=points&year_min=2022&year_max=2022'
dfnew = create_full_df(url, COOKIE_DATA)
display(dfnew)
```
